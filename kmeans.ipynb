```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score
import pandas as pd
import random
```

```python
customers = pd.read_csv('/content/customers.csv')
iris = pd.read_csv('/content/iris.csv')
```

```python
data2d = iris[['sepal_length', 'sepal_width']]
data2d = data2d.to_numpy()
data3d = customers[['YearsEmployed', 'Age', 'Income']]
data3d = data3d.to_numpy()
```

```python
clusters = [3, 4, 5, 6, 7, 8, 9]
```

```python
def clustering(X, er, method):
  score = []
  data = X.copy()
  for i in range(round(len(data)*er)-1):
    ind = random.randint(0, len(data)-1)
    data[ind][0] *= 2
    data[ind][1] *= 2
  for cluster in clusters:
    # Kmeans / Kmedoids
    if method == 1:
      model = KMeans(n_clusters=cluster).fit(data)
      score.append(silhouette_score(data, model.fit_predict(data)))
        
    elif method == 2:
      model = KMedoids(n_clusters=cluster).fit(data)
      score.append(silhouette_score(data, model.fit_predict(data)))

    # 2D / 3D
    if len(X[0]) == 2:
      plt.figure(figsize=(6, 6))
      plt.scatter(data[:,0],data[:,1], c=model.labels_, cmap='rainbow')
      plt.show()
    if len(X[0]) == 3:
      fig = plt.figure(figsize=(6, 6))
      ax = fig.add_subplot(111, projection='3d')
      ax.scatter(data[:,0], data[:,1], data[:,2], c=model.labels_, cmap='rainbow')
      plt.show()
  return score
```

2D K-Means

```python
score = clustering(data2d, 0, 2)
plt.bar(clusters, score)
plt.xlabel('Number of clusters', fontsize=16)
plt.ylabel('S(i)', fontsize=16)
plt.show()
```

3D K-Means

```python
score = clustering(data3d, 0, 2)
plt.bar(clusters, score)
plt.xlabel('Number of clusters', fontsize=16)
plt.ylabel('S(i)', fontsize=16)
plt.show()
```

2D K-Means Error: 0.03

```python
score = clustering(data2d, 0.03, 2)
plt.bar(clusters, score)
plt.xlabel('Number of clusters', fontsize=16)
plt.ylabel('S(i)', fontsize=16)
plt.show()
```
