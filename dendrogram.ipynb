```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd
import math
```

```python
data = pd.read_csv('/content/data.csv')
data = data.to_numpy()
iris = pd.read_csv('/content/iris.csv')
dataIris = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
labelIris = iris['species']
```

```python
linkageTypes = ['single', 'complete', 'average', 'ward']
```

```python
for linkageType in linkageTypes:
  plt.figure(figsize=(12, 6))
  Dendrogram = dendrogram((linkage(data, linkageType)))
  plt.show()
```

```python
labels = list(set(labelIris))
labelNum = []
for i in labelIris:
  labelNum.append(labels.index(i))
```

```python
for linkageType in linkageTypes:
  plt.figure(figsize=(12, 6))
  Dendrogram = dendrogram((linkage(dataIris, linkageType)), labels=labelNum, leaf_font_size=7)
  plt.show()
```
